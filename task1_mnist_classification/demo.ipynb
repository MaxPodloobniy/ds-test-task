### Importing Required Libraries
We start by importing the necessary libraries for loading the MNIST dataset, visualizing the data, and working with our models.


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from MnistClassifier import MnistClassifier
```

### Loading and Visualizing the Data
The MNIST dataset consists of 28x28 grayscale images of handwritten digits (0-9). We will load the dataset and visualize a few examples.


```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to [0, 1]
x_train, x_test = x_train.astype("float32") / 255.0, x_test.astype("float32") / 255.0

def show_images(images, labels, num=10):
    plt.figure(figsize=(10, 5))
    for i in range(num):
        plt.subplot(2, 5, i + 1)
        plt.imshow(images[i], cmap="gray")
        plt.title(f"Label: {labels[i]}")
        plt.axis("off")
    plt.show()

show_images(x_train, y_train, num=10)
```

## Evaluating Different Models
We will initialize and train three different models:
1. Random Forest (rf)
2. Feed-Forward Neural Network (nn)
3. Convolutional Neural Network (cnn)

After training, we will make predictions on a few test images.

### Test Random Forest Model


```python
rf_classifier = MnistClassifier(alg_type='rf')
rf_classifier.train()

print("Testing Random Forest Model:")
predictions = rf_classifier.predict()
indices = range(10)
show_images([x_test[i].reshape(28, 28) for i in indices], [predictions[i] for i in indices])
```

### Test Feed-Forward Neural Network Model


```python
nn_classifier = MnistClassifier(alg_type='nn')
nn_classifier.train()

print("Testing Feed-Forward Neural Network Model:")
predictions = nn_classifier.predict()
predictions = np.argmax(predictions, axis=1)
show_images([x_test[i].reshape(28, 28) for i in indices], [predictions[i] for i in indices])
```

### Test Convolutional Neural Network Model

```python
cnn_classifier = MnistClassifier(alg_type='cnn')
cnn_classifier.train()

print("Testing Convolutional Neural Network Model:")
predictions = cnn_classifier.predict()
predictions = np.argmax(predictions, axis=1)
show_images([x_test[i].reshape(28, 28) for i in indices], [predictions[i] for i in indices])
```

### Conclusion
This notebook demonstrates the classification of handwritten digits using three different approaches: Random Forest, Feed-Forward Neural Network, and Convolutional Neural Network. Each model is implemented as a separate class, following an interface-based design pattern. The predictions showcase the varying performance and capabilities of different algorithms on the MNIST dataset.
